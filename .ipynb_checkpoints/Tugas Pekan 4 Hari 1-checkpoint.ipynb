# Soal 1. Menentukan jenis data categorical atau numeric
Jelaskan data kategorikal (kualitatif) dan data numerik (kuantitatif)



Jawab disini:

- Data kategorikal = adalah Data kategorikal merupakan data yang mencerminkan karakteristik seperti bahasa, jenis kelamin seseorang, warna rambut seseorang, atau nilai numerik bulat yang tidak memiliki makna matematis. Contoh : 1 untuk laki-laki dan 0 untuk perempuan. Data kategorikal tersebut kemudian terbagi atas dua jenis: Nominal dan Ordinal
    1. Data Nominal =  merupakan data yang tidak memiliki order dan berbentuk label tanpa adanya tingkatan perangkingan. Contoh : Jenis kelamin, warna, dsb
    2. Data ordinal =  merupakan data kategorikal yang disusun dalam sebuah order dan diatur menggunakan peringkat-peringkat tertentu. Contohnya adalah status ekonomi, huruf mutu, level pendidikan, dsb
    
- Data numerik = Data numerik terbagi menjadi 2, yaitu kontinyu dan diskrit.
    1. Variabel diskrit merupakan data yang memiliki nilai yang terbata. Contoh : jumlah medali yang dimenangkan oleh para atlet.
    2. Variabel kontinu merupakan jenis data yang berbeda dengan data kategorikal. Data kontinu merupakan data kuantitatif yang dapat diukur, dapat memiliki nilai yang sangat tinggi (infinite). Contoh : Massa bayi, Kecepatan angin, Temperatur, Usia
    - Selain contoh diatas, terdapat 2 tipe data yang masuk ke dalam kategori data kontinyu yaitu data interval dan rasio.
        1.  Data interval merupakan skala dimana dua buah nilai memiliki makna tertentu. Beberapa karakteristik dari data interval adalah sebagai berikut:
            - Pengukuran: data tersebut diukur menggunakan skala interval. Contohnya adalah penggunaan penggaris dimana untuk setiap interval terkecil penanda pertama mengukur jarak temperatur yang sama persis dengan interval terkecil penanda yang lain.
            - Perbedaan interval: pengukuran untuk setiap nilai dari interval data sama persis. Contohnya adalah perbedaan interval antara 10 cm dan 20 cm adalah sama dengan 20 cm dan 30 cm
            - Kalkulasi: Dalam data interval, seseorang dapat menambah-kurangkan angka pada data tersebut, tetapi tidak bisa membagi atau mengalikan data tersebut dengan suatu nilai. 
            - Titik nol: Titik nol dalam data interval dapat berubah-ubah. Contohnya adalah temperatur dapat bernilai negatif
        2. Data rasio merupakan data yang terdefinisi pada data kuantitatif yang memiliki sifat-sifat yang sama dengan data interval. Data rasio memiliki sifat-sifat yang sama dengan data interval. Namun, data rasio memiliki nilai 0 yang absolut. Contoh : pengukuran tinggi badan. Tinggi badan seseorang tidak dapat bernilai negatif (misal -70 cm). Pun halnya dengan massa seseorang yang tidak dapat bernilai negatif (-29 kg).

```python
# read na values and convert to float
df = pd.read_csv('austin_weather.csv', na_values=['na', '-'])
df = df.astype({'DewPointAvgF':'float' , 'HumidityAvgPercent':'float', 'WindAvgMPH':'float'})
df.isnull().sum()
```

```python
# set x, y, color, sizes
x = df['HumidityAvgPercent'].sample(n=1000, random_state=1)
y = df['DewPointAvgF'].sample(n=1000, random_state=1)
colors = df['TempAvgF'].sample(n=1000, random_state=1)
sizes = df['WindAvgMPH'].sample(n=1000, random_state=1) * 20
```

```python
# scatter plot
plt.figure(figsize=(12,8))
plt.scatter(x, y, s=sizes, c=colors, cmap='coolwarm')
plt.xlabel('Humadity Average (%)')
plt.ylabel('Dew Point Average (F)')
plt.title('Austin Weather')
plt.show
```

```python
df = df.fillna({'DewPointAvgF':'ffill' , 'HumidityAvgPercent':'ffill', 'WindAvgMPH':'ffill'})
df.isnull().sum()
```

```python
# scatter plot
plt.figure(figsize=(12,8))
plt.scatter(x, y, s=sizes, c=colors, cmap='coolwarm')
plt.xlabel('Humadity Average (%)')
plt.ylabel('Dew Point Average (F)')
plt.title('Austin Weather')
plt.show
```

```python
# boxplot 'House_Price'
import seaborn as sns

plt.figure(figsize=(12,8))
sns.boxplot(data=df2, x='House_Price')
```

```python
# handling outlier
q1 = df2['House_Price'].quantile(0.25)
q3 = df2['House_Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
print(lower_bound, upper_bound)
```

```python
# filtering data outlier
df2['House_Price'] = df2[(df2['House_Price']>lower_bound) & (df2['House_Price']<upper_bound)]
df2['House_Price']
```

```python
# boxplot after handling outlier
plt.figure(figsize=(12,8))
sns.boxplot(data=df2, x='House_Price')
```
