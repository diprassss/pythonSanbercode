```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(ncols=3, sharex=True, sharey=True, figsize=(20,10))
ax[0].plot(x, tan, color='blue', label='Grafik tanx')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, cos, color='red', label='Grafik cosx')
ax[1].set_title('Grafik cosx')
ax[2].plot(x, sin, color='green', label='Grafik sinx')
ax[2].set_title('Grafik sinx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# main axis
fig, ax = plt.subplots(figsize=(10,9))
ax.plot(x, y)
ax.set_title('Grafik CosX')

# nested axis 1
ins1 = ax.inset_axes([0.9, 0.9, 0.4, 0.3])
ins1.plot(x, y)
ins1.plot(x, y4, color='black')
ins1.set_title('Grafik CosX^4')

# nested axis 2
ins2 = ax.inset_axes([-0.2, 0.9, 0.4, 0.3])
ins2.plot(x, y)
ins2.plot(x, y5, color='yellow')
ins2.set_title('Grafik CosX^5')

# nested axis 3
ins3 = ax.inset_axes([0.9, -0.1, 0.4, 0.3])
ins3.plot(x,y)
ins3.plot(x, y3, color='green')
ins3.set_title('Grafik CosX^3')

# nested axis 4
ins4 = ax.inset_axes([-0.2, -0.1, 0.4, 0.3])
ins4.plot(x,y)
ins4.plot(x, y2, color='red')
ins4.set_title('Grafik CosX^2')
```
